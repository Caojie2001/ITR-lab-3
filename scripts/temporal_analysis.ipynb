```python
import pandas as pd
import matplotlib.pyplot as plt
import glob
import os
```

```python
folder_path = '/Users/caojie/Desktop/MACSS/2024 fall/data mining/week3/kmeans_clustered_data'

all_files = glob.glob(os.path.join(folder_path, "*.csv"))
data_frames = []

for file in all_files:
    df = pd.read_csv(file)
    data_frames.append(df)

data = pd.concat(data_frames, ignore_index=True)

cluster_counts = data.groupby(['year', 'prediction']).size().unstack(fill_value=0)

plt.figure(figsize=(10, 6))
for prediction in cluster_counts.columns:
    plt.plot(cluster_counts.index, cluster_counts[prediction], marker='o', label=f'Cluster {prediction}')

plt.title("Annual Trend of Text Counts by Cluster")
plt.xlabel("Year")
plt.ylabel("Number of Texts")
plt.legend(title="Cluster")
plt.grid()
plt.show()
```
